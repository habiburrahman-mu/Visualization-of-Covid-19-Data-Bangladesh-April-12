```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
import folium
```

```python
df = pd.read_csv('bd-covid19-apr-12.csv')
df.head()
```

```python
df.shape
```

```python
bd_lat = 23.6850
bd_long = 90.3563
sanfran_map = folium.Map(location=[bd_lat, bd_long], zoom_start=8)

for lat, lng, label in zip(df.lat, df.long, df.number):
    folium.features.CircleMarker(
        [lat, lng],
        radius=np.log(label*100),  #define how big you want the circle markers to be
        color='black',
        fill=True,
        popup=str(label),
        fill_color='red',
        fill_opacity=0.8
    ).add_to(sanfran_map)

# show map
sanfran_map
```
